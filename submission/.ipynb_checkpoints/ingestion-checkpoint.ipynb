```python
import os
import requests
import time
import json
from datetime import datetime
import pandas as pd

API_BASE_URL = os.environ.get("API_BASE_URL")
API_USERNAME = os.environ.get("API_USERNAME")
API_PASSWORD = os.environ.get("API_PASSWORD")

BRONZE_PATH = "/home/jovyan/work/output/bronze"
MANIFEST_FILE = f"{BRONZE_PATH}/manifest.json"

os.makedirs(BRONZE_PATH, exist_ok=True)
```

```python
class APIClient:
    def __init__(self):
        self.token = None
        self.get_token()

    def get_token(self):
        resp = requests.post(f"{API_BASE_URL}/api/v1/auth/token", json={
            "username": API_USERNAME,
            "password": API_PASSWORD
        })
        self.token = resp.json()["access_token"]

    def request(self, method, url, **kwargs):
        retries=5
        backoff=2

        for attempt in range (retries):
            headers=kwargs.get("headers",{})
            headers["Authorization"] = f"Bearer {self.token}"
            kwargs["headers"] = headers

            response = requests.request(method, url, **kwargs)

            if response.status_code == 401:
                print("Token expired, refreshing...")
                self.get_token()
                continue

            if response.status_code in [500, 429]:
                print("Retry {attempt+1} for {url}")
                time.sleep(backoff ** attempt)
                continue

            return response

        raise Exception(f"Failed after retries: {url}")
```

```python
def load_manifest():
    if os.path.exists(MANIFEST_FILE):
        with open(MANIFEST_FILE, "r") as f:
            return json.load(f)

    return {}

def save_manifest(manifest):
    with open(MANIFEST_FILE,"w") as f:
        json.dump(manifest,f,indent=2)
    
    
```

```python
def fetch_endpoint(client, endpoint):
    print(f"\n Fetching {endpoint}")

    manifest = load_manifest()
    processed_pages = set(manifest.get(endpoint, []))

    all_data = []
    page = 1

    while True:
        print(f" Fetching page {page}")

        url = f"{API_BASE_URL}/api/v1/data/{endpoint}"
        resp = client.request("GET", url, params={"page": page, "page_size": 1000})

        data = resp.json()

        records = data.get("data", [])
        
        # STOP condition 
        if not records:
            print("No more data, stopping...")
            break

        # Skip only if already processed
        if str(page) in processed_pages:
            print(f"⏭ Skipping page {page}")
            page += 1
            continue

        # Validation
        valid_records = [r for r in records if isinstance(r, dict)]

        # Add metadata
        for r in valid_records:
            r["_ingested_at"] = datetime.utcnow().isoformat()
            r["_source_endpoint"] = endpoint

        all_data.extend(valid_records)

        # Save progress
        manifest.setdefault(endpoint, []).append(str(page))
        save_manifest(manifest)

        print(f"Page {page} done")

        page += 1 

    return all_data
```

```python
def save_to_bronze(endpoint, data):
    if not data:
        print(f"No data for {endpoint}")
        return
    df = pd.DataFrame(data)

    file_path = f"{BRONZE_PATH}/{endpoint}.csv"

    if os.path.exists(file_path):
        df.to_csv(file_path, mode='a', header=False, index=False)

    else:
        df.to_csv(file_path, index=False)

    print(f"Saved {len(df)} records to {file_path}")
```

```python
client = APIClient()

endpoints  = [
    "orders",
    "order_items",
    "customers",
    "products", 
    "sellers",
    "payments"
]

for ep in endpoints:
    data = fetch_endpoint(client,ep)
    save_to_bronze(ep, data)
```
